```python
import numpy as np
from joblib import load

#загружаем матрицу term-documemt и объект CountVectorizer
X = load('term-document.pkl') 
vectorizer = load('vectorizer.pkl') 
X_t = X.T
#просуммируем встречаемости слов по каждому документу
array_freq = np.asarray(X.sum(axis=0)).ravel()
print('Самое частотное слово:')
vectorizer.get_feature_names()[array_freq.argmax()]
```

```python
#у нас не может быть одного редкого слова, их несколько, и и у них у всех частотность равна 1
print('Самые редкие слова: \n')
for word in np.array(vectorizer.get_feature_names())[np.where(array_freq==1)[0]]:
    print(word)
```

```python
print('Набор слов во всех документах коллекции: \n')
#перемножим встречаемости слов по каждому документу и найдем индексы произведения, не равные нулю
#это и есть индексы слов, встречающихся во всех документах
X_arr = np.array(X.toarray(), dtype = np.object)
X_prod = np.prod(X_arr, axis=0)
for word in np.array(vectorizer.get_feature_names())[np.where(X_prod>0)[0]]:
    print(word)
```

```python
#имя Джоуи распарсилось очень странно, и также было замечено, что оно иногда выглядит как Джоуи, а иногда как Джои
heroes = ('моника', 'рэйчел','чендлер','фиби', 'росс', 'джоуя', 'джой')
#для каждого имени просуммируем частоту встречаемости по каждому документу
name_freqs = []
for name in heroes:
    name_freqs.append(np.sum(X_t.getrow(vectorizer.vocabulary_.get(name)).toarray()))
#сложим значения для Джоуи
name_freqs[-2] = name_freqs[-2]+name_freqs[-1]
print('Самый статистически популярный герой: \n')
print(heroes[np.argmax(np.array(name_freqs[:-1]))])
```
